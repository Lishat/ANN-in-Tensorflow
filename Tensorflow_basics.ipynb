```
import numpy as np
import tensorflow as tf
import pandas as pd
```

```
train = pd.read_excel("train1.xlsx")
test = pd.read_excel("test.xlsx")
```

```
x_train = np.array(train.drop(columns=["Class Label"]))
y_train = np.array(pd.get_dummies(train["Class Label"]))
x_test = np.array(test.drop(columns=["Class Label"]))
y_test = np.array(pd.get_dummies(test["Class Label"]))
```

```
def ANN(x_train, hid_units1, hid_units2, y_train, x_test, y_test):
  x = tf.placeholder(tf.float32, [None, 16])
  y = tf.placeholder(tf.float32, [None, 10])
  w1 = tf.Variable(tf.truncated_normal([16, hid_units1], stddev = 0.05), name='w1')
  b1 = tf.Variable(tf.truncated_normal([hid_units1]), name='b1')     
  z1 = tf.matmul(x, w1) + b1 
  z1 = tf.nn.relu(z1)
  w2 = tf.Variable(tf.truncated_normal([hid_units1, hid_units2], stddev = 0.05), name='w2')
  b2 = tf.Variable(tf.truncated_normal([hid_units2]), name='b2')     
  z2 = tf.matmul(z1, w2) + b2
  z2 = tf.nn.relu(z2)
  w3 = tf.Variable(tf.truncated_normal([hid_units2, 10], stddev = 0.05), name='w3')
  b3 = tf.Variable(tf.truncated_normal([10]), name='b3')     
  z3 = tf.matmul(z2, w3) + b3
  a3 = tf.nn.softmax(z3)
  loss = tf.reduce_mean(tf.nn.softmax_cross_entropy_with_logits_v2(logits=z3, labels = y_train))
  optimizer = tf.train.AdamOptimizer(learning_rate=0.0001).minimize(loss)
  accuracy = tf.reduce_mean(tf.cast(tf.equal(tf.argmax(y, 1), tf.argmax(a3, 1)), tf.float32))
  with tf.Session() as sess:
    sess.run(tf.global_variables_initializer())
    for epoch in range(100000):
      _, c = sess.run([optimizer, loss], feed_dict={x:x_train, y:y_train})
      if epoch%10000 == 0:
        print("LOSS after", epoch, "th epoch:", c)
    test_accuracy = sess.run(accuracy, feed_dict={x:x_test, y:y_test})
    print("Test Accuracy:", test_accuracy*100)
```

```
ANN(x_train, 15, 10, y_train, x_test, y_test)
```
